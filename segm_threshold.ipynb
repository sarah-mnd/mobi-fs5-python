```python
import numpy as np
import matplotlib.pyplot as plt
```

**[Task 1.2 (a).]()** Use `plt.imread` to load an image:

```python
X = plt.imread('data/NIH3T3/im/dna-0.png')
```

**[Task 1.2 (b).]()** Show the image:

```python
plt.figure(figsize=(15,8))
plt.imshow(X)
plt.colorbar()
```

**[Task 1.2 (d).]()** Binarize the image by applying an intensity threshold:

```python
T = 0.1
X >= T #True wenn Pixelwerte >= T, sonst False
X_binarized = (X >= T).astype(int) # True wird 1, False wird 0 (int fr integer)
```

**[Task 1.2 (e).]()** Load the ground truth:

```python
plt.clf()
plt.figure(figsize=(15,8))
plt.title('X_binarized')
plt.imshow(X_binarized, cmap='gray')
plt.colorbar()
plt.clf()
ground_truth = plt.imread('data/NIH3T3/gt/0.png')
plt.figure(figsize=(10,4))
plt.title('ground truth')
plt.imshow(ground_truth, cmap='gray')
```

Compute and print the Dice coefficient:

```python
X_bin_bool = X_binarized.astype(bool)
#print(X_bin_bool.dtype)
ground_truth_bool = ground_truth.astype(bool)
#print(X_bool.dtype)
G = ground_truth_bool > 0
H = X_bin_bool > 0
G_and_H = G * H
Dice = 2 * (G_and_H.sum()) / (G.sum()+ H.sum())
print("Dice Coefficient = ", Dice)
print("Dice Coefficient = 1 -> perfect overlap")
print("Dice Coefficient = 0 -> no overlap")
```
